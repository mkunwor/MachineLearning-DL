```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=[10,11,12,13,
         14,15,16,17,18,20,22,24,25,26,30,34,32,33,40,45,50,50,55,60,90,120,180]

```

```python
sns.histplot(dataset,color='r',kde='true')
```

```python
import seaborn as sns
sns.boxplot(dataset)
```

```python
#z score
outliers=[]

#function to take data as argument and return possible  outliers
def detect_outliers(data):
  threshold=3  #3 std deviation
  mean=np.mean(data)
  std=np.std(data)
  for i in data:
    z_score=(i-mean)/std  #element in which std
    if np.abs(z_score)>threshold:
      outliers.append(i)
  return outliers
```

```python
detect_outliers(dataset)
```

```python
dataset=sorted(dataset)
q1,q2,q3=np.percentile(dataset,[25,50,75])
print(q1,q2,q3)
```

```python

iqr=q3-q1
print(iqr)
```

```python

#find lower fence and higer fence
lower_fence=q1-(1.5*iqr)
higher_fence=q3+(1.5*iqr)
print(lower_fence,higher_fence)
```

```python
outliers=[]
for item in dataset:
  if item<lower_fence or item>higher_fence:
    outliers.append(item)
print(outliers)
```

```python
import seaborn as sns
sns.boxplot(dataset)
```

```python
for item in outliers:
    dataset.remove(item)
```
