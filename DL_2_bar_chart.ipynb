```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMAZON','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label="Revenue",color='b')
plt.xticks(xpos,company)
plt.xlabel("software companies")
plt.ylabel("revenue(Bln)")
plt.title("US technology stocks")
plt.legend()
```

```python
profit=[40,2,34,12]
netprofit=[50,1.5,40,15]
```

```python
plt.bar(xpos-0.2,revenue,width=0.2,label="Revenue")
plt.bar(xpos,profit,width=0.2,label="Profits")
plt.bar(xpos+0.2,netprofit,width=0.2,label='NetProfit')
plt.xticks(xpos,company)
plt.ylabel("Revenuew(bln)")
plt.title("Us technology stocks")
plt.legend()
```

```python
plt.barh(xpos,revenue,label="Revenue",color='b')
plt.xticks(xpos,company)
plt.xlabel("software companies")
plt.ylabel("revenue(Bln)")
plt.title("US technology stocks")
plt.legend()
```
