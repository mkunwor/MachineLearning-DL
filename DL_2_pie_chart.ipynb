```python
import matplotlib.pyplot as plt
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','Car','Other']
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%2.3f%%',radius=1.5)
```

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%2.3f%%',radius=1.5, explode=[0,0,0,1,0.2])
```

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%2.3f%%',radius=1.5, explode=[0,0,0,0.1,0.2],counterclock=False,startangle=60)
```
